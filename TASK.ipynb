Part 5: Small Challenge

Build a small calculator that:

Asks the user for two numbers usint input() function and an operation (+, -, *, /)

Performs the calculation and prints the result

Keeps asking until the user types “exit”

```python
# Example numbers and operation
num1 = 5
num2 = 3
operation = '+'  


if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 == 0:
        result = "Cannot divide by zero!"
    else:
        result = num1 / num2
else:
    result = "Invalid operation!"

print(f"The result of {num1} {operation} {num2} is: {result}")
```

```python
import pandas as pd
```

```python
df_raw= pd.read_excel(r"D:\DATA ANALYSIS\SalesData.xlsx",header=None)
```

```python
print(df_raw.info())
```

```python
df=df_raw
print(df.shape)
```

```python
df=df_raw[0].str.split("," , expand=True)
print(df.head())
```

```python
df.columns=df.iloc[1]
```

```python
df=df[2:]
```

```python
df.head()
```

```python
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
missing_values=df.isnull().sum()
```

```python
print(missing_values)
```

```python

df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').fillna(0).astype(int)
```

```python
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce').fillna(0).astype(int)
```

```python
df
```

Part 2: Data Analysis

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0).astype(int)
```

```python
df['Sales']=df['Quantity']*df['UnitPrice']
total_sales=df['Sales'].sum()

print("Total Sales Amount=",total_sales)
```

```python
Sales_per_Region=df.groupby('Region')['Sales'].sum()
```

```python
print(Sales_per_Region.reset_index())
```

```python
Quantity_per_category=df.groupby('Category')['Quantity'].sum()
print(Quantity_per_category.reset_index())
```

```python
sales_per_product=df.groupby('Product')['Sales'].sum()
print(sales_per_product.reset_index())
```

```python
highest_selling_product = sales_per_product.idxmax()
highest_sales_value = sales_per_product.max()
```

```python
print("Highest Selling Product:", highest_selling_product)
print("Total Revenue:", highest_sales_value)
```
